# Empowering Aspiring Investors: Loan Approval for Real Estate


Historically, real estate investment has been a strong wealth-building strategy, with a historical increase in property values over time. However, entering real estate has relatively large fixed costs (upfront costs), creating a high barrier to entry. In most cases, recent graduates wait quite a few years to purchase a real estate property, often requiring a loan in the form of a mortgage, to purchase a property. 

For aspiring real estate investors, particularly those aged 20 to 30, a vital question they must ask themselves is: Will I be approvaled for a loan?

This project addresses this question using data sourced from the Federal Financial Institutions Examinaiton Council, which is a government entity that regulates and collects data from financial instituions such as banks and credit unions. I will investigate the factors that impact loan approval, and develop a predictive model that can predict if an individual will be approved for a loan. This will help support my financial goals, as well as those seeking to take informed steps towards their real estate investment goals.

# Adjusting the Training Data

#### This first model has high accuracy, but this is due to the disproportionate data: class 0 has significantly fewer observations than class 1. The model yields a "high" accuracy mainly by predicting more of class 1, rather than effectively using the model features. Therefore, undersampling may be an effective way to mitigate this issue by reducing the number of observations in class 1.

```python
# creating approved and reject dataframes
df1_approved = df1[df1['Approval'] == 1].reset_index(drop=True)
df1_rejected = df1[df1['Approval'] == 0].reset_index(drop=True)
```

```python
# looking at approved counts
df1_approved['Approval'].count()
```

```python
# looking at rejected counts
df1_rejected['Approval'].count()
```

```python
df1_approved
```

```python
df1_rejected
```

### Applying Undersampling Method

```python
from sklearn.utils import resample
```

```python
# making the sample size equivalent to size of underperesented class (class 0)
sample_size = len(df1_rejected)
```

```python
# take the random sample
approved_sample = resample(df1_approved, replace=False, n_samples=sample_size)
rejected_sample = resample(df1_rejected, replace=False, n_samples=sample_size)
```

```python
approved_sample
```

```python
rejected_sample
```

```python
# combining both samples to 1 datframe
random_sample = pd.concat([approved_sample, rejected_sample]).reset_index(drop = True)
```

```python
# combining both samples to 1 datframe
random_sample = pd.concat([approved_sample, rejected_sample]).reset_index(drop = True)
```

```python
# creating a copy to split to features and labels
sample = random_sample.copy()
```
